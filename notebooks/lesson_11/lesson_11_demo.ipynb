# Introduction to Data Visualization

1. Plotting a quantity: bar plot
2. Plotting a continuous function: line plot
3. Plotting relationships: scatter plots
4. Plotting distributions: box plot & histograms

## 1. Bar plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate data for a bar chart
categories = ['Category A', 'Category B', 'Category C']
values = [25, 40, 15]
```

```python
plt.title('Bar Chart')

plt.bar(categories, values, color='grey', edgecolor='black') #params are x axis and then y axis
plt.xlabel('Categories')
plt.ylabel('Values')
plt.xticks(rotation=45)

plt.show()
```

## 2. Line plot

```python
# Generate data for a line plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
plt.title('Line plot')

plt.plot(
    x, 
    y, 
    color='red', 
    linestyle='--', 
    marker='o',
    markerfacecolor='black',
    markeredgecolor='black',
    markersize=4
)

plt.show()
```

## 3. Scatter plot

```python
# Generate data for a scatter plot
x = np.random.rand(100)
y = 2 * x + np.random.randn(100)
```

```python
plt.title('Scatter plot')
plt.scatter(x, y, color='black', s=10)

plt.show()
```

## 4. Distribution plots

### 4.1 Histogram

```python
data = np.random.normal(0, 1, 1000)
data
```

```python
plt.title('Histogram')
plt.hist(data, bins=30, color='gray', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Count')
plt.show()
```

### 4.2 Boxplot

```python
# Generate data for a box plot
data = [np.random.normal(0, 1, 100), np.random.normal(0, 1.5, 100), np.random.normal(0, 2, 100)]
```

```python
plt.title('Boxplot')
plt.boxplot(data)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

n = 1000
x = np.random.uniform(0, 10, n)

y1_linear = 0.5 * x + 10
y2_linear = 2.5 * x + 0.1

y1_noise = y1_linear + np.random.normal(0, 1, n)
y2_noise = y2_linear + np.random.normal(0, 1, n)

df = pd.DataFrame({
    'x': x,
    'y1_linear': y1_linear,
    'y2_linear': y2_linear,
    'y1_noise': y1_noise,
    'y2_noise': y2_noise
})

df.head()
```

```python
plt.scatter(df['x'], df['y1_noise'], s=10, alpha=0.2, label='Feature 1')
plt.scatter(df['x'], df['y2_noise'], s=10, alpha=0.2, label='Feature 2')

plt.plot(df['x'], df['y1_linear'])
plt.plot(df['x'], df['y2_linear'])

plt.legend()
plt.show()
```

### Multiple subplots

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 6))
axs = axs.flatten()

axs[0].scatter(df['x'], df['y1_noise'])
axs[2].hist(df['y2_noise'])

plt.savefig('incomplete_multiplot.png')
plt.show()
```

## Seaborn

```python
import seaborn as sns
```

```python
tips_df = sns.load_dataset('tips')
tips_df.head()
```

### Line plot

```python
sns.lineplot(x='day', y='total_bill', data=tips_df, hue='smoker')
plt.title('Tips by day')
plt.show()
```

### Scatter Plot

```python
sns.scatterplot(x='total_bill', y='tip', data=tips_df, hue='smoker')
plt.title('Total bill vs tip')
plt.show()
```

### Distribution plots

##### Histogram

```python
sns.histplot(tips_df['total_bill'], kde=True)
plt.show()
```

### KDE

```python
sns.kdeplot(tips_df['total_bill'])
plt.show()
```

```python
sns.pairplot(tips_df, hue='sex')
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips_df, hue='sex')
plt.show()
```
